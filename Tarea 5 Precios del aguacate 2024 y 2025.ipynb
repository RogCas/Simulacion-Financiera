#Tarea 5 Precios del aguacate en 2023,2024,2025

En esta tarea es exactamente lo mismo que la tarea 2

Pero en este caso tenemos los datos de los aguacates que tendran un incremento de precio, que se distribuye de manera uniforme en los años 2024 U(0,0.046) y 2025 U(0,0.0315)

Entonces solo es multiplicar la media de los precios que calculamos antes por 1+ un numero aleatorio de las uniformes de 2024 y 2025 y en base a eso calcular las nuevas ganancias.




```python
import random
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd
n=500  #numero de simulaciones


padv=.52 #probabilidad adversa
pfav=.48 #probabilidad favorable
pCS=.26 #probabilidad Crisis Severa
pCL=.29 #probabilidad Crisis Leve
pMO=.27 #probabilidad Moderado
pCR=.18 #probabilidad Crecimiento

media = 1000
desviacion_estandar = 160

probabilidad=[padv*pCS,pfav*pCS,padv*pCL,pfav*pCL,padv*pMO,pfav*pMO,padv*pCR,pfav*pCR]

pacumulada=[sum(probabilidad[:i+1]) for i in range(len(probabilidad))]

gastos = [48, 39, 43, 35, 40, 30, 35, 23]

simGastos=[]
listaU1=[]
for i in range(0,n):
  u1=random.random()
  listaU1.append(u1)
  for i in range(0, len(pacumulada)+1):
      if u1<=pacumulada[i]:
        simGastos.append(gastos[i])
        break

mediaGastos=(sum(simGastos)/len(simGastos))
gastosReales=mediaGastos*1000000


simProduccion=[]
listaU2=[]
for i in range(0,n):
  u2 = random.random()
  listaU2.append(u2)
  produccion = norm.ppf(u2, loc=media, scale=desviacion_estandar)
  simProduccion.append(produccion)

mediaProduccion=sum(simProduccion)/len(simProduccion)

simPrecios=[]
listaU3=[]
for i in range(0,n):
  u3=random.random()
  listaU3.append(u3)
  precio=100*u3+100
  simPrecios.append(precio)

mediaPrecios=sum(simPrecios)/len(simPrecios)



sim2024=0.046*random.random()
sim2025=0.0315*random.random()
precio2024=[]
precio2025=[]
sim2024a=[]
sim2025a=[]

for i in range(0,n):
  x=simPrecios[i] * (1+sim2024)
  precio2024.append(x)
  y=simPrecios[i] * (1+sim2025)
  precio2025.append(y)

  sim2024a.append(sim2024)

  sim2025a.append(sim2025)

mediaPrecios24=sum(precio2024)/len(precio2024)
mediaPrecios25=sum(precio2025)/len(precio2025)

#Hacemos las operaciones
ventas=mediaProduccion*mediaPrecios*1000
ganancias=ventas-gastosReales

ventas24=mediaProduccion*mediaPrecios24*1000
ganancias24=ventas24-gastosReales

ventas25=mediaProduccion*mediaPrecios25*1000
ganancias25=ventas25-gastosReales


#Hacemos que tengan coma para que sea mas facil la visualizacion
precioscoma = "{:,}".format(mediaPrecios)
ventascoma = "{:,}".format(ventas)
gananciascoma = "{:,}".format(ganancias)
gastoscoma = "{:,}".format(gastosReales)

precioscoma24 = "{:,}".format(mediaPrecios24)
ventascoma24 = "{:,}".format(ventas24)
gananciascoma24 = "{:,}".format(ganancias24)

precioscoma25 = "{:,}".format(mediaPrecios25)
ventascoma25 = "{:,}".format(ventas25)
gananciascoma25 = "{:,}".format(ganancias25)


#Imprimimos los datos para 2023"
print("\n Año 2023")
print("\nPrecios simulados(previstas) con ",n,"simulaciones : ", precioscoma)
print("\nVentas simuladas(previstas) con ",n,"simulaciones : ", ventascoma)
print("\nGastos simulados(previstos) con",n,"simulaciones: ", gastoscoma)
print("\nLas ganacias simuladas (previstas) con",n,"simulaciones: ", gananciascoma)
print("\n")

#Imprimimos los datos para 2024"
print("\n Año 2024")
print("\nPrecios simulados(previstas) con ",n,"simulaciones : ", precioscoma24)
print("\nVentas simuladas(previstas) con ",n,"simulaciones : ", ventascoma24)
print("\nGastos simulados(previstos) con",n,"simulaciones: ", gastoscoma)
print("\nLas ganacias simuladas (previstas) con",n,"simulaciones: ", gananciascoma24)
print("\n")

#Imprimimos los datos para 2025"
print("\n Año 2025")
print("\nPrecios simulados(previstas) con ",n,"simulaciones : ", precioscoma25)
print("\nVentas simuladas(previstas) con ",n,"simulaciones : ", ventascoma25)
print("\nGastos simulados(previstos) con",n,"simulaciones: ", gastoscoma)
print("\nLas ganacias simuladas (previstas) con",n,"simulaciones: ", gananciascoma25)
print("\n")

```
