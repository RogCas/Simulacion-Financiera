##Depreciacion Computadora

Nuestros datos nos dicen que el año actual vale 70 mil pesos, en el siguiente año valdra 49 mil pesos y en 5 años valdra 10 mil pesos, este precio lo escogimos en la clase.

Proponemos un modelo exponencial
y = A * exp(a * x)

Y resolvemos

Ae^(a0)	=	$70,000.00

A	=	$70,000.00

Ae^(a)	=	$49,000.00

70000e^(a)	=	$49,000.00

e^(a)	=	49,000/70,000

a	=	ln(49,000/70,000)

a	=	ln(7/10)

alpha	=	-0.356674944

Redondeamos a 4 digitos

alpha	=	-0.356600000

y	=	70,000e^(-.3566)


Ya que tenemos la funcion completamos para los años del 0 al 5

Al completar sacamos la diferencia de los valores que ya teniamos

Y con base a eso sacamos otra exponencial, pero esta sera para el errro

Error

Ae^(a)	=	$3.6724

Ae^(a5)	=	$1,769.3094

A	=	3.6724/e^(a)

(3.6724/e^(a))*e^(5a)	=	$1,769.3094

(3.6724)*e^(4a)	=	$1,769.3094

e^(4a)	=	1,769.3094/3.6724

4a	=	ln(1,769.3094/3.6724)

alpha	=	$1.54437480

alpha	=	$1.54430000

A	=	3.6724/e^(1.5443)

A	=	0.783915064409736

A	=	0.7839

y	=	0.7839*e^1.5443

Y sacamos los errores que nos falten




```python
#Codigo Computadora
#Redondeado a 4 decimales

import math
#Funcion para poner $ y coma en los miles
def coma(x):
    return "${:,.2f}".format(x)

n=5 #numero de años para simular

an0=70000
a0=coma(an0)

an1=49000
a1=coma(an1)

an5=10000
a5=coma(an5)

dineroSim=[]

for i in range(n+1):
  dineroSim.append(an0*math.exp(-.3566*i))
# redondear a 4 decimales

dineroD = [int(num * 10**4) / 10**4 for num in dineroSim]
# darle formato
dinero4 = [coma(num) for num in dineroD]

#Calculamos el error
a0E=dineroD[0]-an0
a1E=dineroD[1]-an1
a5E=dineroD[5]-an5
error=[]
for i in range(n+1):
  error.append(.7839*math.exp(1.5443*i))

error[0]=a0E
error[1]=a1E
error[5]=a5E

```

```python
from tabulate import tabulate


# Datos para la tabla
data = [
    ["Año", "Dinero", "Dinero Simulado", "Error"],
    [0, a0, dinero4[0], error[0]],
    [1, a1, dinero4[1], error[1]],
    [2, "", dinero4[2], error[2]],
    [3, "", dinero4[3], error[3]],
    [4, "", dinero4[4], error[4]],
    [5, a5, dinero4[5], error[5]]
]

# Imprimir la tabla
print(tabulate(data, tablefmt="plain"))
```

```python
#Usando Solo los puntos
import matplotlib.pyplot as plt

# Datos
X = [i for i in range(n+1)]
Y = dineroD

# Graficar los datos
plt.plot(X, Y, marker='o', linestyle='-')
plt.xlabel('Años')
plt.ylabel('Dinero')
plt.title('Depreciacion de la computadora durante 5 años usando solo 5 puntos')
plt.grid(True)
plt.show()


#Usando la grafica
import numpy as np
import matplotlib.pyplot as plt

# Definir la función exponencial
def funcion_exponencial(x, A, b):
    return A * np.exp(b * x)

# Parámetros de la función
A = an0
b = -.3566

# Rango de valores de x
x_values = np.linspace(0, 5, 100)  # de 0 a 5 con 100 puntos

# Calcular los valores de y usando la función exponencial
y_values = funcion_exponencial(x_values, A, b)

# Graficar la función
plt.plot(x_values, y_values, label=f'Ae^bx, A={A}, b={b}')
plt.xlabel('Años')
plt.ylabel('Dinero')
plt.title('Depreciacion de la computadora durante 5 años usando la exponencial')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Codigo Departamento
#Redondeado a 4 decimales

import math
#Funcion para poner $ y coma en los miles
def coma(x):
    return "${:,.2f}".format(x)

n=6 #numero de años para simular

an0=1490000
a0=coma(an0)

an5=1810000
a5=coma(an5)

an6=1900000
a6=coma(an6)

dineroSim=[]

for i in range(n+1):
  dineroSim.append(an0*math.exp(.0389*i))
# redondear a 4 decimales

dineroD = [int(num * 10**4) / 10**4 for num in dineroSim]
# darle formato
dinero4 = [coma(num) for num in dineroD]

#Calculamos el error
a0E=-dineroD[0]+an0
a5E=-dineroD[5]+an5
a6E=-dineroD[6]+an6
error=[]
for i in range(n+1):
  error.append(0.00000000029188*math.exp(5.2949*i))

error[0]=a0E
error[5]=a5E
error[6]=a6E

#le damos formato

#error = [coma(num) for num in error]
```
