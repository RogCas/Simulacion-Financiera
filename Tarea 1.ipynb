```python
import random
import matplotlib.pyplot as plt

n=10000
a=0.5
b=3
ejeX=[]
ejeY=[]
for numero in range(1, n+1):
  aleatorio = random.random()
  x=((b-a)*aleatorio)+a
  ejeX.append(x)
  gx=x*x
  ejeY.append(gx)
plt.scatter(ejeX, ejeY, label='Puntos', color='blue', marker='o')

plt.xlabel('X')
plt.ylabel('X^2')
plt.title('X^2')



# Agregar leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

n=10000

ejeX=[]#aceleracion
ejeY=[]#masa
ejeZ=[]#Fuerza
for numero in range(1, n+1):
  masa=random.uniform(18,25)
  tiempo=2*random.random()
  velocidad=np.random.normal(loc=80, scale=19)
  aceleracion=velocidad/tiempo
  fuerza=masa*aceleracion
  ejeX.append(aceleracion)
  ejeY.append(masa)
  ejeZ.append(fuerza)

#Grafico aceleracion y masa
plt.scatter(ejeX, ejeY, label='Fuerza=m*a', color='blue', marker='o')
plt.xlabel('Aceleracion')
plt.ylabel('Masa')
plt.title('Fuerza')
plt.legend()
plt.show()

#Grafico aceleracion y Fuerza
plt.scatter(ejeX, ejeZ, label='Fuerza vs a', color='blue', marker='o')
plt.xlabel('Aceleracion')
plt.ylabel('Fuerza')
plt.title('Fuerza')
plt.legend()
plt.show()

#Grafico masa y Fuerza
plt.scatter(ejeY, ejeZ, label='Fuerza vs m', color='blue', marker='o')
plt.xlabel('Masa')
plt.ylabel('Fuerza')
plt.title('Fuerza')
plt.legend()
plt.show()
```
